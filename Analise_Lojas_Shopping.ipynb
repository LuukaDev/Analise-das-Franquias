# Realizar uma análise detalhada do desempenho das franquias localizadas em shoppings, com o objetivo de identificar oportunidades para aumentar o faturamento geral da rede.

- A análise buscará mapear padrões, tendências e variáveis que influenciam os resultados das lojas. Com base nesses dados, serão gerados insights estratégicos que contribuam para a otimização do desempenho e para a tomada de decisões mais precisas e eficazes.

Importação da base de dados do shopping: os dados foram extraídos diretamente do Google Drive.

```python
import pandas as Pd
Dados_Shopping = Pd.read_excel("/content/drive/MyDrive/B) Portifolio/Analise de dados/Vendas.xlsx")
display(Dados_Shopping)
```

Análise do Faturamento Total das Lojas

- Antes de aprofundar na análise individual das franquias, vamos observar o faturamento total de cada shopping. Esse dado serve como base para entender o potencial de consumo de cada loja



```python
Faturamento_total = Dados_Shopping['Valor Final'].sum()
display('Faturamento', Faturamento_total)
```

Conforme os dados analisados, o faturamento total consolidado de todos os shoppings é de R$ 55.895,00 no período considerado.

Com base no faturamento, é possível realizar uma análise comparativa entre as lojas para identificar quais estão com melhor desempenho e quais apresentam resultados abaixo da média. Essa avaliação permite levantar hipóteses sobre os motivos das variações e sugerir estratégias para melhorar o desempenho das unidades com menor faturamento.

```python
Lojas_Comparação = Dados_Shopping[['ID Loja','Produto','Valor Final']].groupby(['ID Loja','Produto','Valor Final']).sum()
display(Lojas_Comparação)
```

A partir da análise dos dados extraídos, foi possível identificar um ponto bastante chamativo: a loja localizada no Shopping Iguatemi Campinas é responsável por aproximadamente 65,46% do faturamento total. Esse dado evidencia um desempenho significativamente superior em relação às demais unidades, o que pode indicar boas práticas a serem replicadas ou características específicas desse ponto que contribuem para seu sucesso.

```python
Iguatemi_Rede = Dados_Shopping[Dados_Shopping['ID Loja'] == 'Iguatemi Campinas']
display(Iguatemi_Rede)
```

Na unidade do Shopping Iguatemi Campinas, nos dias 5, 24 e 25 de dezembro de 2019, foi registrado um volume elevado de vendas do produto Bermuda Liso, superando significativamente o desempenho das outras redes nesse período.


```python
import plotly.express as px

grafico_Shop = px.histogram(Dados_Shopping, x='ID Loja', y='Valor Final',
                      title='Faturamento por Loja',
                      labels={'ID Loja': 'Loja', 'Valor Final': 'Faturamento'},
                      color='ID Loja')
grafico_Shop.show()
```

```python
grafico_Shop = px.histogram(Dados_Shopping, x='Produto', y='Valor Final',
                      title='Faturamento por Loja',
                      labels={'ID Loja': 'Loja', 'Valor Final': 'Faturamento'},
                      color='ID Loja')
grafico_Shop.show()
```
